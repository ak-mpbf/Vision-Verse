```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# I have used nearest neighbour interpolation method for downsampling images.
def sampling(image, factor):
  img1=cv2.imread(image,0)
  [m,n]=img1.shape
  img2=np.zeros((m//factor,n//factor), dtype=int)
  for i in range(0,m,factor):
    for j in range(0,n,factor):
      try:
        img2[i//factor][j//factor]=img1[i][j]
      except IndexError:
        pass
  return img2
```

```python
input='input.jpg' #Virat Kohli fan :)
print("Original Image:")
input_img=cv2.imread(input,0)
plt.imshow(input_img)
```

```python
img2x=sampling(input,2)
print("Down Sampled Images:")
print("2x:")
plt.imshow(img2x)
```

```python
img4x=sampling(input,4)
print("4x:")
plt.imshow(img4x)
```

```python
def quantization(image, level):
  img3=cv2.imread(image,0)
  step=256//level
  img4=(img3//step)*step
  return img4
```

```python
input='input.jpg'
print("Original Image:")
input_img=cv2.imread(input,0)
plt.imshow(input_img)
```

```python
imgl4=quantization(input,4)
print("Quantized Images:")
print("Level=4")
plt.imshow(imgl4)
```

```python
imgl8=quantization(input,8)
print("Level=8")
plt.imshow(imgl8)
```

```python
imgl16=quantization(input,16)
print("Level=16")
plt.imshow(imgl16)
```
