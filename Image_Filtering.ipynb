```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
from google.colab.patches import cv2_imshow
```

```python
i1='/content/lowContrast.png'
i2='/content/original.jpg'
image1=cv2.imread(i1)
image2=cv2.imread(i2)
```

**GAUSSIAN FILTER**

```python
image1_g = cv2.GaussianBlur(image1, (5, 5), 0)
cv2_imshow(image1_g)
image2_g = cv2.GaussianBlur(image2, (5, 5), 0)
cv2_imshow(image2_g)
```

**MEDIAN FILTER**

```python
image1_m = cv2.medianBlur(image1, 5)
cv2_imshow(image1_m)
image2_m = cv2.medianBlur(image2, 5)
cv2_imshow(image2_m)
```

**LAPLACIAN FILTER**

```python
image1_gray = cv2.cvtColor(image1_g, cv2.COLOR_BGR2GRAY)
image1_l = cv2.Laplacian(image1_gray, cv2.CV_16S, ksize=3)
plt.imshow(image1_l)
```

```python
image2_gray = cv2.cvtColor(image2_g, cv2.COLOR_BGR2GRAY)
image2_l = cv2.Laplacian(image2_gray, cv2.CV_16S, ksize=3)
plt.imshow(image2_l)
```

**BILATERAL FILTER**

```python
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image1_b = cv2.bilateralFilter(image1_gray, d=5, sigmaColor=60, sigmaSpace=60)
cv2_imshow(image1_b)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
image2_b = cv2.bilateralFilter(image2_gray, d=5, sigmaColor=60, sigmaSpace=60)
cv2_imshow(image2_b)
```

**BOX FILTER**

```python
image1_box = cv2.boxFilter(image1, 0, (5,5))
cv2_imshow(image1_box)
image2_box = cv2.boxFilter(image2, 0, (5,5))
cv2_imshow(image2_box)
```

**INFERENCE :-** Bilateral Filter gives the best results as it preserves the edges.
